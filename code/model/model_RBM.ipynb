```python
'''
Utility class with some helper functions
'''

import pandas as pd
import numpy as np
import random
import os

class Util(object):

    def read_data(self, folder = '/content/drive/MyDrive/scrap/'):
        '''
        Function to read data required to
        build the recommender system
        '''
        print("Reading the data")
        review = pd.read_csv(folder + 'reviews_data.csv')
        attract = pd.read_csv(folder + 'attractions.csv')
        return review, attract

    def preprocess(self, review, attract):
        '''
        Preprocess data for feeding into the network
        '''
        print("Preprocessing the dataset")

        data = review.merge(attract, on='attraction_id')
        data = data[['user_id','attraction_id','rating_x']]
        data_aggregated = data.groupby(['user_id', 'attraction_id']).agg({'rating_x': 'mean'}).reset_index()
        user_rating_df = data_aggregated.pivot(index='user_id', columns='attraction_id', values='rating_x')
        norm_user_rating = user_rating_df.fillna(0) / 5.0
        trX = norm_user_rating.values

        return trX


    def free_energy(self, v_sample, W, vb, hb):
        '''
        Function to compute the free energy
        '''
        print(v_sample.shape)
        print(w.shape)
        wx_b = np.dot(v_sample, W) + hb
        vbias_term = np.dot(v_sample, vb)
        hidden_term = np.sum(np.log(1 + np.exp(wx_b)), axis = 1)
        return -hidden_term - vbias_term
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

class RBM:
    def __init__(self, visible_units, hidden_units):
        self.visible_units = visible_units
        self.hidden_units = hidden_units
        self.w = tf.Variable(tf.random.normal([visible_units, hidden_units], 0.01), dtype=tf.float32)
        self.vb = tf.Variable(tf.zeros([visible_units]), dtype=tf.float32)
        self.hb = tf.Variable(tf.zeros([hidden_units]), dtype=tf.float32)
        self.errors = []
        self.weights = []

    def hidden_layer(self, v0_states):
        v0_states_reshaped = tf.reshape(v0_states, [-1, self.visible_units])
        h0_prob = tf.nn.sigmoid(tf.matmul(v0_states_reshaped, self.w) + self.hb)
        h0_state = tf.nn.relu(tf.sign(h0_prob - tf.random.uniform(tf.shape(h0_prob))))
        return h0_state

    def reconstructed_output(self, h0_state):
        v1_prob = tf.nn.sigmoid(tf.matmul(h0_state, tf.transpose(self.w)) + self.vb)
        v1_state = tf.nn.relu(tf.sign(v1_prob - tf.random.uniform(tf.shape(v1_prob))))
        return v1_state[0]

    def error(self, v0, v1):
        return tf.reduce_mean(tf.square(v0 - v1))

    def train(self, trX, epochs=10, batchsize=500, k=1, alpha=0.1):
        train_ds = tf.data.Dataset.from_tensor_slices(np.float32(trX)).batch(batchsize)
        j = 0
        for epoch in range(epochs):
            batch_number = 0
            for batch_x in train_ds:
                j += 1
                for i_sample in range(len(batch_x)):
                    v0_state = batch_x[i_sample]
                    h0_state = self.hidden_layer(v0_state)
                    v1_state = self.reconstructed_output(h0_state)
                    h1_state = self.hidden_layer(v1_state)

                    delta_w = tf.matmul(tf.transpose([v0_state]), h0_state) - tf.matmul(tf.transpose([v1_state]), h1_state)

                    self.w = self.w + alpha * delta_w
                    self.vb = self.vb + alpha * tf.reduce_mean(v0_state - v1_state, 0)
                    self.hb = self.hb + alpha * tf.reduce_mean(h0_state - h1_state, 0)

                    if i_sample == len(batch_x) - 1:
                        err = self.error(batch_x[i_sample], v1_state)
                        self.errors.append(err)
                        self.weights.append(self.w.numpy())

                        print('iteration: %d' % (j),
                              'Epoch: %d' % (epoch + 1),
                              "batch #: %i" % batch_number,
                              "sample #: %i" % (i_sample + 1),
                              'reconstruction error: %f' % err)
                    batch_number += 1

        # Plotting the errors
        plt.plot(self.errors)
        plt.ylabel('Error')
        plt.xlabel('Iterations')
        plt.show()
        self.save()




    def save(self, filename='word'):
        if not os.path.exists('rbm_models'):
            os.mkdir('rbm_models')
        filename = 'rbm_models/'+filename
        if not os.path.exists(filename):
            os.mkdir(filename)
        np.save(filename+'/w.npy', self.w)
        np.save(filename+'/vb.npy', self.vb)
        np.save(filename+'/hb.npy', self.hb)


# Utilisation de la classe RBM
visible_units = len(trx[0])
hidden_units = 128  # Ajustez selon vos besoins
rbm = RBM(visible_units, hidden_units)
rbm.train(trx)
```
