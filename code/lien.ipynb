```python
def get_proxy(link):
    url = "https://www.sslproxies.org/"
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    proxies_list = list(map(lambda x: x[0]+':'+x[1], list(zip(map(lambda x: x.text, soup.findAll('td')[::8]), map(lambda x: x.text, soup.findAll('td')[1::8])))))
    while 1:
        try:
            selected_ip = choice(proxies_list)
            proxy = {'https': selected_ip, 'http': selected_ip}
            headers = {'User-Agent': user_agent.random}
            print('Using proxy:{}'.format(proxy))
            r = requests.request('get', link, proxies=proxy, headers=headers, timeout=5)
            break
        except:
            pass
        
    return proxy
```

```python
def clean_text(text):
    text = text.strip().strip('-')
    text = text.lower()
    text = '_'.join(text.split(' '))
    return text
```

```python
def scrapDetail(url):
    item = {}
   
    userAgent = user_agent.random
    Firefox_options = webdriver.EdgeOptions()
    Firefox_options.add_argument(f'user-agent={userAgent}')
    driver = webdriver.Edge(options=Firefox_options)
    driver.get(url)
    time.sleep(random.uniform(5,10))
    driver.set_window_size(1080, 800)

    detail_soup = BeautifulSoup(driver.page_source, 'html.parser')
    card = detail_soup.find('div', {"div", 'yNgTB A'})
 
    list_img = detail_soup.find('div', attrs={"class":"qrDQR"})
    img_item_list = [a['src'] for a in list_img.find_all('img')]
    
    item['image_detail'] = img_item_list
    text = card.find('div', {'class':'fIrGe _T bgMZj'})
    
    item['text'] = getItem(text)
    dure = card.find_next('div', {'class':"_c"})
   
    item['duree'] = getItem(dure)
    rest = card.find('div', {'class':'xPwwX'})
    item['restriction'] = getItem(rest)

    score = detail_soup.find('div', {'class':'yFKLG'})

    note = score.find('div', {'class':'biGQs _P fiohW hzzSG uuBRH'})
    item['note'] = getItem(note)
    review = score.find('span', {'class':'biGQs _P pZUbB KxBGd'})
    item['review'] = getItem(review)
    
    cles = score.find_all('div', {'class':'nVhPm Pv o q Vm W Cj PX'})
    val = score.find_all('div', {'class':'nVhPm Pv o q Vm W Cj w'})

    cles_val = [getItem(a) for a in cles]
    val_val = [getItem(a) for a in val]

   
    driver.quit()
    return {**item, **dict(zip(cles_val, val_val))}

  

   
```

```python
def scrapAllReviewsGet(url, attraction_id):
    
    items = []
       
    userAgent = user_agent.random
    Firefox_options = webdriver.EdgeOptions()
    Firefox_options.add_argument(f'user-agent={userAgent}')
    driver = webdriver.Edge(options=Firefox_options)
    driver.get(url)
    #driver.set_window_size(1080, 800)
    time.sleep(random.uniform(5,17))
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    #card_reviews = soup.find('div', {"class", 'LbPSX'})
    #reviews = card_reviews.find_all('div', {"class", '_c'})

    reviews = soup.select('div[data-automation="tab"] div.C div._c') #div_country.find_all('div', attrs={"class":"KCGqk _T Cj"}) 

    for review in reviews:
        item = {}
        photo_user = review.find('div', {"class", 'MLvbw f u'}) #
        photo_user = photo_user.find('a', {'class', 'BMQDV _F Gv wSSLS SwZTJ'})

       

        try:
            photo_user = photo_user.find('img')
            item['image_user'] = photo_user['src'] 
        except:
            item['image_user'] = ''
             

        
        user = review.find('div', {"class", 'zpDvc Zb'})
        item['name_user'] = getItem(user.find('span', {'class', 'biGQs _P fiohW fOtGX'}))
        item['info_user'] = getItem(user.find('div', {'class', 'JINyA'}))
        item['type_review'] = getItem(review.find('span', {'class': 'yCeTE'}))
        item['date_review'] = getItem(review.find('div', {'class': 'biGQs _P pZUbB ncFvv osNWb'})).split(' ', 1)[1]
        item['text_review'] = getItem(review.find('div', {'class': 'fIrGe _T bgMZj'}))
        span_score = review.find('svg', {'class': 'UctUV d H0'})  #UctUV d H0
        item['score_user'] = float(clean_text(span_score.get('aria-label')).replace('_of_5_bubbles',''))
        item['attraction_id'] = attraction_id

        #print(item)


        items.append(item)
    # appuye sur le bouton suivant et relancer la fonction pour scraper
    
    
    print(f'nombre data scrapé: ', {len(items)})
    driver.quit()

    return items


```

```python
def createFileCoppyData(items, chemin_fichier_csv = 'data.csv'):
    # cette fonction va prendre le tableau de dictionnaire et
    # enregistrer chaque ligne dans le tableau

    items_data = pd.DataFrame(items)
    items_data.to_csv(chemin_fichier_csv, mode='a', header=not os.path.exists(chemin_fichier_csv),  index=False)
```

```python
import requests

def get_openstreetmap_image_link(latitude, longitude, zoom=18, width=400, height=300):
    base_url = "https://nominatim.openstreetmap.org/search"
    
    params = {
        "format": "json",
        "lat": latitude,
        "lon": longitude,
        "zoom": zoom,
        "width": width,
        "height": height,
    }

    response = requests.get(base_url, params=params)

    print(f"https://www.openstreetmap.org/cbox?lat={latitude}&lon={longitude}&zoom={zoom}&width={width}&height={height}")

    if response.status_code == 200:
        # Obtenez le premier résultat de la réponse JSON
        data = response.json()
        if data:
            place = data[0]
            # Construisez l'URL de l'image à partir des informations obtenues
            image_url = f"https://www.openstreetmap.org/cbox?lat={latitude}&lon={longitude}&zoom={zoom}&width={width}&height={height}"
            return image_url
        else:
            print("Aucun résultat trouvé.")
            return None
    else:
        print(f"Erreur lors de la requête à l'API Nominatim. Code d'erreur: {response.status_code}")
        return None

def main():
    latitude = input("Entrez la latitude : ")
    longitude = input("Entrez la longitude : ")

    image_url = get_openstreetmap_image_link(latitude, longitude)

    if image_url:
        print(f"Lien de l'image OpenStreetMap : {image_url}")
    else:
        print("Impossible d'obtenir le lien de l'image.")

if __name__ == "__main__":
    main()
```
